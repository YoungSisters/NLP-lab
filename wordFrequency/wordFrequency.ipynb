```python
# 불용어 제거
import nltk

nltk.download('all')
nltk.download('wordnet')
nltk.download('stopwords')

from nltk.corpus import stopwords
stop_word_eng = set(stopwords.words('english'))
line = [i for i in line if i not in stop_word_eng]
```

```python
for i in range(len(stopwords.words('english'))):
  print(stopwords.words('english')[i], end=' ')
  if (i%10==9):
    print()
```

```python
import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.stem.porter import PorterStemmer
ps_stemmer = PorterStemmer()
token = RegexpTokenizer('[\w]+')
result2 = [token.tokenize(i) for i in line]
middle_result= [r for i in result2 for r in i]
final_result = [ps_stemmer.stem(i) for i in middle_result if not i in stop_word_eng] # 불용어 제거
print(final_result)
```

```python
### 표제어 추출 사용 시 ###
import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
token = RegexpTokenizer('[\w]+')
result_pre_lem = [token.tokenize(i) for i in line]
middle_pre_lem= [r for i in result_pre_lem for r in i]
final_lem = [lemmatizer.lemmatize(i) for i in middle_pre_lem if not i in stop_word_eng] # 불용어 제거

print(final_lem)

for i in range(len(final_lem)):
  print(final_lem[i], end=' ')
  if (i%10==9):
    print()
```

```python
###텍스트에서 많이 나온 단어###
import pandas as pd
english = pd.Series(final_lem).value_counts().head(10) #.head(10)
print("TOP 10 WORDS")
print(english)
```
